```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading dataset
athletes=pd.read_csv('athlete_events.csv')
regions=pd.read_csv('noc_regions.csv')
```

```python
athletes.head()
```

```python
regions.head()
```

```python
#Joining the dataframes
athletes_df=pd.merge(athletes,regions,on='NOC',how='left')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
#Making column names consistent
athletes_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
athletes_df.head()
```

```python
athletes_df.info()
```

```python
#Statistical summary
athletes_df.describe()
```

```python
#checking for null values
nan_values=athletes_df.isna()
nan_columns=nan_values.any()
nan_columns
```

```python
athletes_df.isnull().sum()
```

```python
#Participation from India
athletes_df[athletes_df.Team=='India'].head()
```

```python
#Participation from Japan
athletes_df[athletes_df.Team=='Japan'].head()
```

```python
#Top 10 coutries Participation 
top_ten_countries=athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_ten_countries
```

```python
#Plot for the countries
plt.figure(figsize=(12,6))
plt.title('Overall participation by country')
sns.barplot(x=top_ten_countries.index,y=top_ten_countries,palette='Set2')
plt.show()
```

```python
#Age distribution of the participants
plt.figure(figsize=(12,6))
plt.title('Age distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age,bins=np.arange(10,80,2),color='orange',edgecolor='white')
plt.show()
```

```python
#Winter olympics sports
winter_sports=athletes_df[athletes_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
#Summer olympics sports
summer_sports=athletes_df[athletes_df.Season=='Summer'].Sport.unique()
summer_sports
```

```python
#Male and Female participants
gender_counts=athletes_df.Sex.value_counts()
gender_counts
```

```python
#Total medals
athletes_df.Medal.value_counts()
```

```python
#Total number of female athletes in each olympics
female_participants=athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants
```

```python
womenOlympics=female_participants=athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')]
```

```python
#Women participation
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womenOlympics,palette='Spectral')
plt.title('Women participation')
plt.show()
```

```python
#Gold medal athletes
goldMedals=athletes_df[athletes_df.Medal=='Gold']
goldMedals.head()
```

```python
#values that are not NaN
goldMedals=goldMedals[np.isfinite(goldMedals.Age)]
```

```python
#Gold beyond 60 Years old
goldMedals['ID'][goldMedals.Age > 60].count()
```

```python
#Sporting events with Gold medal above 60 years
sporting_events=goldMedals['Sport'][goldMedals.Age>60]
sporting_events
```

```python
#plot for Sporting events with Gold medal above 60 years
plt.figure(figsize=(10,5))
sns.countplot(sporting_events)
plt.title('Gold medal won by Athletes above 60 years')
plt.show()
```

```python
#Gold medals from each country

goldMedals.Region.value_counts().reset_index(name='Medals').head()
```

```python
#Rio Olympics
max_year=athletes_df.Year.max()
team_names=athletes_df[(athletes_df.Year==max_year) & (athletes_df.Medal=='Gold')].Team
team_names.value_counts().head(10)
```

```python
#plot for coutry wise gold medals in Rio Olympics
sns.barplot(x=team_names.value_counts().head(20),y=team_names.value_counts().head(20).index)
plt.xlabel('Country wise gold medals in the Rio Olympics')
plt.ylabel(None)
plt.show()
```
